```python
X = df[['CRIM','NOX','RM','AGE','TAX']]
X
```

```python
#행렬곱
X_t.dot(X)
```

```python
#공분산 행렬
X_t@X
```

```python
XtX = X_t@X
```

```python
np.linalg.det(XtX)
```

```python
XtX_inv = np.linalg.inv(XtX)
print(XtX_inv)
```

```python
XtX @ XtX_inv
```
